```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df =pd.read_excel('(预处理后)2019-2023Data.xlsx',sheet_name='2019-2023Team',header=0)
TeamData = np.array(df)
df =pd.read_excel('(预处理后)2019-2023Data.xlsx',sheet_name='2019-2023Player',header=0)
PlayerData = np.array(df)
```

```python
Team2019 = []
Team2020 = []
Team2021 = []
Team2022 = []
Team2023 = []
Player2019 = []
Player2020 = []
Player2021 = []
Player2022 = []
Player2023 = []

for i in range(50):
    Team2019.append(TeamData[i])
    Team2020.append(TeamData[i+50])
    Team2021.append(TeamData[i+98])
    Team2022.append(TeamData[i+150])
    Team2023.append(TeamData[i+200])
    Player2019.append(PlayerData[i])
    Player2020.append(PlayerData[i+50])
    Player2021.append(PlayerData[i+100])
    Player2022.append(PlayerData[i+150])
    Player2023.append(PlayerData[i+200])

print('Top50 Team in 2019:',Team2019)
print('Top50 Team in 2020:',Team2020)
print('Top50 Team in 2021:',Team2021)
print('Top50 Team in 2022:',Team2022)
print('Top50 Team in 2023:',Team2023)
print('Top50 Player in 2019:',Player2019)
print('Top50 Player in 2020:',Player2020)
print('Top50 Player in 2021:',Player2021)
print('Top50 Player in 2022:',Player2022)
print('Top50 Player in 2023:',Player2023)

```

```python
df2 = pd.read_excel('(预处理后)2019-2023Data.xlsx',sheet_name='2019-2023Team',header=0)
Team = df2.iloc[:,0]
Team1st = df2.iloc[:,1]
Team2nd = df2.iloc[:,2]
Team3rd_4th = df2.iloc[:,3]
Totalwinnings = df2.iloc[:,4]
df3 = pd.read_excel('(预处理后)2019-2023Data.xlsx',sheet_name='2019-2023Player',header=0)
Player = df3.iloc[:,0]
Player1st = df3.iloc[:,1]
Player2nd = df3.iloc[:,2]
Player3rd_4th = df3.iloc[:,3]
PlayerS_Tier = df3.iloc[:,4]
Earnings = df3.iloc[:,5]
```

```python
Teamsum = []
Playersum = []
for i in range(250):
    Teamsum.append(Team1st[i] + Team2nd[i] + Team3rd_4th[i])
    Playersum.append(Player1st[i] + Player2nd[i] + Player3rd_4th[i] + PlayerS_Tier[i])

for year in range(2019,2024):
    print(str(year),Team,Teamsum,Totalwinnings)
    print(str(year),Player,Playersum,Earnings)
```

```python
for year in range(2019,2024):
    TotalPrize = 0
    for i in range(50*(year-2019),50*(year-2019)+50):
        TotalPrize += Totalwinnings[i]
    print(str(year)+'全年比赛总奖金额度为$',TotalPrize)
```

```python
#查看最年轻选手与最年长选手
da =pd.read_excel('(MySQL)Apex Legends Global Teams and Players.xlsx',sheet_name='Youngest',header=0)
Youngest = np.array(da)
print(Youngest)
Yp = da.iloc[:,0]
Yage = da.iloc[:,1]
da =pd.read_excel('(MySQL)Apex Legends Global Teams and Players.xlsx',sheet_name='Oldest',header=0)
Oldest = np.array(da)
print(Oldest)
Op = da.iloc[:,0]
Oage = da.iloc[:,1]
sumy = 0
sumo = 0
for i in range(20):
    sumy += Yage[i]
    sumo += Oage[i]
print("最年轻选手的平均年龄为:",sumy/len(Yage),"岁")    
print("最年长选手的平均年龄为:",sumo/len(Oage),"岁")  

x = []
y = []
for i in range(40):
    if i < 20:
        x.append(Yp[i])
        y.append(Yage[i])
    else:
        x.append(Op[i-20])    
        y.append(Oage[i-20])
#年龄图可视化
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(40,10))
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=20)
plt.xlabel("选手")
plt.ylabel("年龄")    
plt.show()
```

```python
dt = pd.read_excel('(MySQL)Apex Legends Global Teams and Players.xlsx',sheet_name='TeamOverall',header=0)
TeamO = dt.iloc[0:10,0].values
dt = pd.read_excel('(MySQL)Apex Legends Global Teams and Players.xlsx',sheet_name='PlayerOverall',header=0)
PlayerO = dt.iloc[0:10,0].values
#数据可视化
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(15,5),dpi=200)
data2019 = [0]*10
data2020 = [0]*10
data2021 = [0]*10
data2022 = [0]*10
data2023 = [0]*10
for i in range(10):      
    for j in range(50):
        if Team2019[j][0] == TeamO[i]:
            data2019[i] = Team2019[j][4]

        if Team2020[j][0] == TeamO[i]:
            data2020[i] = Team2020[j][4]

        if Team2021[j][0] == TeamO[i]:
            data2021[i] = Team2021[j][4]
            
        if Team2022[j][0] == TeamO[i]:
            data2022[i] = Team2022[j][4]
            
        if Team2023[j][0] == TeamO[i]:
            data2023[i] = Team2023[j][4]

bottom1 = data2019
bottom2 = np.sum([data2019,data2020],axis=0).tolist()
bottom3 = np.sum([bottom2,data2021],axis=0).tolist()
bottom4 = np.sum([bottom3,data2022],axis=0).tolist()
plt.bar(TeamO,data2019,label="2019Totalwinnings",color='blue')
plt.bar(TeamO,data2020,label='2020Totalwinnings',bottom=bottom1,color='green')
plt.bar(TeamO,data2021,label='2021Totalwinnings',bottom=bottom2,color='yellow')
plt.bar(TeamO,data2022,label='2022Totalwinnings',bottom=bottom3,color='red')
plt.bar(TeamO,data2023,label='2023Totalwinnings',bottom=bottom4,color='pink')
plt.legend()
plt.ylabel("TotalWinnings($)")
plt.show()
print(PlayerO)
```
